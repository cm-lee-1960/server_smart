```python
qs = CenterManageArea.objects.all()
for cma in qs:
    centerName = cma.bonbu[:2]
    if centerName == '충남' or centerName == '충북': centerName = '충청'
    print(centerName)
    qs = Center.objects.filter(centerName=centerName)
    if qs.exists():
        center = qs[0]
        print(center)
        cma.center = center
        cma.save()
```
